```python
####################################
## Sitzung 9
## 04.06.2013

# random sidenote: pypdf für PDF Bearbeitung in Python

# reuter news archive für artikel, schon fertig aufbearbeitet, aber kostenpflichtig

# Web of Science abchecken

%matplotlib inline

import nltk
from nltk.corpus import names
import random


###################################
## Naive Bayes Classifier
## Geschlechteridentifikation

# Überblick verschaffen
names = nltk.corpus.names
print names.fileids()

# Z.B.: Bedingte Häufigkeiten:
cfd = nltk.ConditionalFreqDist(
          (fileid, name[-1])
          for fileid in names.fileids()
          for name in names.words(fileid))
cfd.plot()

# Namen die in beiden Gruppen vorkommen:
beide = [w for w in names.words('male.txt')
         if w in names.words('female.txt')]

#print beide

# Alternativ (schneller):
beide = sorted(set(names.words('male.txt')) &
               set(names.words('female.txt')))

print beide 

## 1. Trainingskorpus erstellen
names = ([(name, 'male') for name in names.words('male.txt')] +
         [(name, 'female') for name in names.words('female.txt')])

print names[:10]
random.shuffle(names)
print names[:10]
```

```python
## 2. Eigenschaften definieren

# Funktionen zur Extraktion formulieren:

def gender_features(word):
    return {'last_letter': word[-1]}

## 3. Eigenschaften extrahieren
# Format: Liste von Tuplen (Eigenschafts-Diktionär, Label) 
featuresets = [(gender_features(name), gender) for (name, gender) in names]


print featuresets[:5]

# Ein Test-Set zurückhalten für spätere
# Überprüfung des Klassifizierers
train_set, test_set = featuresets[500:], featuresets[:500]

## 4. Modell trainieren
classifier = nltk.NaiveBayesClassifier.train(train_set)

## 5. Klassifizieren

print classifier.classify(gender_features('Neo'))

print classifier.classify(gender_features('Trinity'))
```

```python
# Ab wann funktioniert es nicht mehr?
# Ausprobieren!

## 6. Modell überprüfen

# Einfache Tests:

print nltk.classify.accuracy(classifier, test_set)

classifier.show_most_informative_features(20)

# Besser: Error analysis (s.u.)

###################################
## Übung:
## Weitere Eigenschaften hinzufügen


###################################
##Exkurs I
## Bei großen Datensätzen:
from nltk.classify import apply_features

# Erlaubt die Konstruktion von Generatoren
# Gleiches Prinzip wie bei xrange()
train_set = apply_features(gender_features, names[500:])
test_set = apply_features(gender_features, names[:500])
```

```python
##################################
## 'Overfitting' des Models
## Zuviele Eiegenschaften bei kleinem Korpus
## verschlechtern das Modell

# Natural Language Toolkit: code_gender_features_overfitting
def gender_features2(name):
    features = {}
    features["firstletter"] = name[0].lower()
    features["lastletter"] = name[-1].lower()
    for letter in 'abcdefghijklmnopqrstuvwxyz':
        features["count(%s)" % letter] = name.lower().count(letter)
        features["has(%s)" % letter] = (letter in name.lower())
    return features

featuresets = [(gender_features2(n), g) for (n,g) in names]
train_set, test_set = featuresets[500:], featuresets[:500]
classifier = nltk.NaiveBayesClassifier.train(train_set)
print nltk.classify.accuracy(classifier, test_set)
```

```python
#################################
## Error analysis

## 1. Development set:
# zerlegt in dev-test und training set
train_names = names[1500:]
devtest_names = names[500:1500]

##test_names = names[:500]
# Für die abschließenden Tests muss ein seperates
# test set gebildet werden!

# Vorgehensweise für error analysis:
# Eigenschaften extrahieren
train_set = [(gender_features(n), g) for (n,g) in train_names]
devtest_set = [(gender_features(n), g) for (n,g) in devtest_names]

# Für später
##test_set = [(gender_features(n), g) for (n,g) in test_names]

# Modell mit training set trainieren
classifier = nltk.NaiveBayesClassifier.train(train_set)

# Genauigkeit
print nltk.classify.accuracy(classifier, devtest_set)

# Das Modell auf dev-test anwenden und die Fehler, die
# bei der Klassifikation gemacht werden bestimmen

errors = []
for (name, tag) in devtest_names:
    guess = classifier.classify(gender_features(name))
    if guess != tag:
        errors.append( (tag, guess, name) )

# Fehler ausgeben:
for (tag, guess, name) in sorted(errors):
    print 'correct=%-8s guess=%-8s name=%-30s' % (tag, guess, name)

## Wie können wir die Extraktion der Eigenschaften anpassen?
```

```python
###############################
## Erweiterte Überprüfung

## Confusion Matrix
# 2 Listen:
# Richtige Verteilung aus Ursprungskorpus und
# Verteilung nach Klassifizierungsmodell

##################################
## Entwurf einer Crossvalidation Funktion
## für Naive-Bayes-Klassifikation
## Kann ziemlich lange dauern!

import nltk
import random

def cross_validation(featuresets, n_of_testsets, runs=5):
    for run in runs:
        random.shuffle(featuresets)
        train_set = featuresets[n_of_testsets:]
        test_set = featuresets[:n_of_testsets]
        classifier = nltk.NaiveBayesClassifier.train(train_set)
        print 'Run %s Accuracy: %s' % run nltk.classify.accuracy(classifier, test_set)
```
