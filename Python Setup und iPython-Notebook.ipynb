# Python - die ersten Schritte

Das nachfolgende Tutorial ist für Personen geschrieben, die mit Python noch keine Erfahrungen gemacht haben und einen unkomplizierten Einstieg in die Installation und Anwendung der freien Programmiersprache suchen. Dabei stellt die Kombination der hier verwendeten Hilfsmittel *Anaconda* und *iPython* nur eine von vielen Möglichkeiten zum Umgang mit Python dar. Nach der Installation von Python folgen ein paar Anwendungsbeispiele, die zeigen sollen, was mit ein paar Zeilen Code gemacht werden kann. Am Ende findet sich noch eine kleine Sammlung an weiterführenden Links zum Erlernen der Sprache.

Vor der ersten Nutzung muss Python zunächst korrekt eingerichtet werden. Python ist zwar auf vielen Betriebssystemen schon vorinstalliert, meistens werden aber neben dem Basisprogramm noch einige zusätzliche Pakete (quasi eine Form von Stata ado-Files, nur sehr viel komplexer) benötigt.
Die Installation jedes einzelnen Pakets ist vor allem für Windows-User  am Anfang umständlich und zeitaufwendig, weshalb sich die Nutzung einer Sammeldistribution empfiehlt. Ein sehr gutes Gesamtpaket stellt  <a href="https://store.continuum.io/cshop/anaconda/" target="_blank">Anaconda</a> dar.

## Anaconda

Anaconda beinhaltet die wichtigsten Python-Pakete in einem Installer für alle gängigen Betriebssysteme und legt automatisch <a href="http://de.wikipedia.org/wiki/Umgebungsvariable" target="_blank">Umgebungsvariablen</a> an, sodass Python einfach über die Eingabe des Kommandos *python* in der <a href="http://de.wikipedia.org/wiki/Betriebssystem-Shell "target="_blank">System-Shell</a> (in Windows die *"Eingabeaufforderung"*) gestartet werden kann.

![Alt Text](https://dl.dropboxusercontent.com/u/22352379/Shell%20Screenshot.png)

#### Auswahl der Version

Bis dato existeren zwei unabhängig voneinander entwickelte Python Versionen: Python 2 und Python 3. Python 3 bietet einige neue Features und beseitigt Schwächen der alten Version (z.B. Probleme bei der <a href ="http://de.wikipedia.org/wiki/Unicode"target="_blank">Unicode</a>-Kompatibilität). 
Allerdings wurden viele der wichtigsten Python-Pakete in Version 2 geschrieben und eine Konvertierung zu Python 3 ist insbesondere bei umfangreichen Paketen nicht ohne Weiteres möglich. Es empfiehlt sich deshalb zumindest momentan noch mit Python 2 zu arbeiten und dementsprechend die Python 2 64-Bit Version von Anaconda zu installieren.

#### Nach der Installation

Insofern man die Einrichtung von Umgebungsvariablen bei der Installation erlaubt hat, können Python und dazugehörige Pakete über die Systemshell aufgerufen werden. Anaconda bietet über den folgenden Befehl eine einfache Möglichkeit, sämtliche Pakete auf den aktuellsten Stand zu bringen:

```markdown 
conda update conda
 ```

Für den Fall, dass ein Paket benötigt wird, welches nicht in der Anaconda-Distribution enthalten ist, kann dieses  über den Paketmanager <a href= "https://pypi.python.org/pypi/pip"target="_blank">pip</a> nachinstalliert werden. Hierfür muss lediglich die Bezeichnung bekannt sein unter welcher das Paket registriert wurde.
Der folgende Befehl installiert den Python-Klon des bekannten *R* Grafikpakets <a href ="http://blog.yhathq.com/posts/ggplot-for-python.html"target="_blank">ggplot</a>:

```markdown
pip install ggplot
```

Obwohl Python ohne Probleme auch über die Shell gesteuert werden könnte, bieten sich sowohl für das Lernen von Python als auch für die fortgeschrittene Anwendung spezielle Entwicklungsumgebungen (IDE) an. Eine solche <a href ="http://de.wikipedia.org/wiki/Integrierte_Entwicklungsumgebung"target="_blank">IDE</a> ist auch in der Standardinstallation von Python enthalten: *Python-IDLE*. Die Entwicklungsumgebung, die Python von den meisten anderen Programmiersprachen absetzt und einen der Hauptfaktoren für die zunehmende <a href ="http://cacm.acm.org/blogs/blog-cacm/176450-python-is-now-the-most-popular-introductory-teaching-language-at-top-us-universities/fulltext"target="_blank">Popularität</a> von Python darstellt, ist das auf *iPython* basierende <a href ="http://ipython.org/notebook.html"target="_blank">iPython-Notebook</a>.

## iPython-Notebook

Das iPython Notebook ist eine browserbasierte, interaktive Programmierumgebung für Python (dieses Tutorial stellt ein Beispiel dar). Innerhalb des iPython Notebooks können in Echtzeit Code ausgeführt, Text geschrieben, Plots dargestellt werden und vieles mehr. Die Notebooks können anschließend gespeichert und in verschiedene Formate (<a href ="http://ipython.org/ipython-doc/1/interactive/nbconvert.html"target="_blank">z.B. Latex und anschließend PDF</a>) konvertiert werden. Mit minimalen Anpassungen ist es so möglich, publikationsfähige Dokumente ohne großen Aufwand zu erstellen.



Dabei reichen die Darstellungsmöglichkeiten soweit, dass mit iPython-Notebooks schon vollständige, interaktive Bücher geschrieben wurden. Ein Beispiel ist die Einführung in die Bayes-Statistik <a href ="http://nbviewer.ipython.org/github/CamDavidsonPilon/Probabilistic-Programming-and-Bayesian-Methods-for-Hackers/blob/master/Prologue/Prologue.ipynb"target="_blank">Probabilistic Programming and
Bayesian Methods for Hackers</a>. Der Link führt zu einer über den <a href = "http://nbviewer.ipython.org/"target="_blank">Notebook-Viewer</a> generierten HTML-Version des Notebooks. Über den Viewer bereitgestellte Notebooks können aber auch heruntergeladen und in die eigene iPython-Notebook-Umgebung eingefügt werden, wodurch der im Buch verwendete Code eigenhändig verändert werden kann.

iPython inklusive Notebook sind in der Anaconda-Installation bereits integriert. Der folgende Shell-Befehl startet das iPython-Notebook:

    ipython notebook

````python
# Fibonacci-Folge als Funktion

def fibonacci(z):
    x,y = 0,1
    l = []
    while x < z:
        l.append(x)
        x, y = y, x + y     
    return l

print fibonacci(10000)
````

````python
import statsmodels.formula.api as sm

url = "http://vincentarelbundock.github.com/Rdatasets/csv/HistData/Guerry.csv"
df = pd.read_csv(url)
df = df[['Lottery', 'Literacy', 'Wealth', 'Region']].dropna()
df.head()
````

````python
# Textanalyse

import urllib
import nltk
import xml.etree.ElementTree as ET
tedtalks = urllib.urlretrieve("https://www.dropbox.com/s/9bl9b3m4yu726zq/ted_transcripts_en_jan2014.xml?dl=1", "ted.xml")
root = ET.parse('ted.xml').getroot()
text = [file[1].text for file in root]
string = "".join(text).lower()
tokens = nltk.Text(nltk.word_tokenize(string))
tokens.concordance("innovation", lines=15)
````

````python
# Widgets, mit denen Programmcode über Browserelemente verändert werden kann
# Die Browserelemente funktionieren nur, wenn das Notebook über iPython geladen
# wird und nicht über die HTML-Darstellung des Notebook-Viewers.

from IPython.html.widgets import interact
%matplotlib inline
import matplotlib.pyplot as plt
import networkx as nx

def random_lobster(n, m, k, p):
    return nx.random_lobster(n, p, p / m)

def powerlaw_cluster(n, m, k, p):
    return nx.powerlaw_cluster_graph(n, m, p)

def erdos_renyi(n, m, k, p):
    return nx.erdos_renyi_graph(n, p)

def newman_watts_strogatz(n, m, k, p):
    return nx.newman_watts_strogatz_graph(n, k, p)

def plot_random_graph(n, m, k, p, generator):
    g = generator(n, m, k, p)
    nx.draw(g)
    plt.show()
    
interact(plot_random_graph, n=(2,30), m=(1,10), k=(1,10), p=(0.0, 1.0, 0.001),
        generator={'lobster': random_lobster,
                   'power law': powerlaw_cluster,
                   'Newman-Watts-Strogatz': newman_watts_strogatz,
                   u'Erdős-Rényi': erdos_renyi,
                   });
````
